# Prisoners Dilemma

## Section 1: Nonfixed graph, standard model

We import the main python file:

```python
from Iterated_prisoners_dilemma import *
```

Now define the main variables.
$$
D - \text{This is a dictionary of agents as keys and their number as values.}\\
Game\_Matrix\_P1,Game\_Matrix\_P2 - \text{These are the payoff matrices for P1 and P2 (note not the transpose of P2).}\\
iterations - \text{The number of iterations to play the graph for.}\\
birth\_death\_rate - \text{How many agents to kill/create each iteration.}\\
rounds - \text{How many rounds to play pairs of agents against each other for.}\\
type\_of\_graph - \text{The type of graph to play the game on.}\\
m - \text{The Barabasi-Albert/Watts-Strogatz variable.}
$$

```python
D = {cheat_agent:40,nice_agent:40,random_agent:40,grudge_agent:40}
Game_Matrix_P1 = [[2,-1],[3,0]]
Game_Matrix_P2 = [[2,-1],[3,0]]
iterations = 100
birth_death_rate = 5
rounds = 10
type_of_graph = nx.barabasi_albert_graph # Includes support for nx.barabasi_albert_graph/watts_strogatz_graph/
m = 10                                   #                         complete_graph/cycle_graph
```

Run the experiment:

```python
%%time
A = multiple_tournament(D,Game_Matrix_P1,Game_Matrix_P2,rounds,birth_death_rate,iterations,m,type_of_graph)
```

Plot the results:

```python
%pylab inline
plot_PD(A,D,iterations)
```
